```python
import numpy as np
```

Đọc file

```python
reader = open('/content/s1.txt')
text = reader.read()
```

```python
lines = text.strip().split('\n')
print (len(lines))
```

```python
arr = []
for line in lines:
  line = line.strip().split()
  line_int = []
  for item in line:
    line_int.append(int(item))
  arr.append(line_int)
  # break
```

```python
arr_np = np.array(arr)
```

```python
print (arr_np.shape)
```

```python
arr_np
```

```python
import matplotlib.pyplot as plt
```

```python
arr_np
```

```python
plt.figure(figsize=(20,10))
plt.scatter(arr_np[:,0], arr_np[:,1])
plt.show()
```

https://scikit-learn.org/stable/modules/generated/sklearn.cluster.AgglomerativeClustering.html

```python
from sklearn.cluster import AgglomerativeClustering

clustering = AgglomerativeClustering(n_clusters=None,distance_threshold=400000).fit(arr_np)
```

```python
n_clusters = max(clustering.labels_)
n_clusters
```

```python
arr_visualize = []
for i in range(max(clustering.labels_)):
  arr = []
  for j, label in enumerate(clustering.labels_):
    if label == i:
      arr.append(arr_np[j])
  arr = np.array(arr)
  arr_visualize.append(arr)
```

```python
plt.figure(figsize=(20,10))
for i in range(max(clustering.labels_)):
  plt.scatter(arr_visualize[i][:,0], arr_visualize[i][:,1])
plt.show()
```

https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html?highlight=k%20means#sklearn.cluster.KMeans

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=16, random_state=0).fit(arr_np)
```

```python
n_clusters = kmeans.n_clusters
```

```python
arr_visualize = []
for i in range(n_clusters):
  arr = []
  for j, label in enumerate(kmeans.labels_):
    if label == i:
      arr.append(arr_np[j])
  arr = np.array(arr)
  arr_visualize.append(arr)
```

```python
plt.figure(figsize=(20,10))
for i in range(n_clusters):
  plt.scatter(arr_visualize[i][:,0], arr_visualize[i][:,1])

centers = kmeans.cluster_centers_
plt.scatter(centers[:,0],centers[:,1],marker='o',s=1000)
plt.show()
```
