```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

# Model1
- Hiểu cấu trúc 1 lớp model của pytorch
- Hiểu nguyên tắc làm việc của một số layer cơ bản
- Hiểu cấu trúc của tensor
- Thay đổi code để chạy với dữ liệu đầu vào có 3 kênh?
- x = x.flatten(start_dim = 1), dự đoán tham số start_dim làm gì?

```python
class MyModel1(nn.Module):
    def __init__(self):
        super(MyModel1, self).__init__()
        # 28x28x1 => 26x26x32
        self.conv1 = nn.Conv2d(in_channels=1, out_channels=32, kernel_size=3)
        self.d1 = nn.Linear(26 * 26 * 32, 128)
        self.d2 = nn.Linear(128, 10)
    def forward(self, x):
        # 32x1x28x28 => 32x32x26x26
        x = self.conv1(x)
        x = F.relu(x)
            
        # flatten => 8 x (32*26*26) 
        print('before flatten: ', x.shape)
        x = x.flatten(start_dim = 1)

        # 32 x (32*26*26) => 32x128
        x = self.d1(x)
        x = F.relu(x)
        
        # logits => 32x10
        logits = self.d2(x)
        print(logits)
        out = F.softmax(logits, dim=1)
        return out
```

```python
model = MyModel1()
x = torch.Tensor(4, 1, 28, 28) # batch, channel, rows, cols
out=model(x)
print(out.shape)
print(out)
```

# Thử nghiệm với một số lớp khác
- Cố gắng hiểu từng tham số

```python
class MyModel3(nn.Module):
    def __init__(self):
        super(MyModel3, self).__init__()

        # 28x28x1 => 26x26x32
        self.conv1 = nn.Conv2d(in_channels=3, out_channels=32, kernel_size=3)
        self.pool1 = nn.MaxPool2d(kernel_size=2, stride=2)
        self.conv2 = nn.Conv2d(in_channels=32, out_channels=64, kernel_size=3)      
        self.pool2 = nn.AdaptiveAvgPool2d((1,1))
        self.d1 = nn.Linear(64, 128)
        self.d2 = nn.Linear(128, 10)

    def forward(self, x):
        # 32x1x28x28 => 32x32x26x26
        x = self.conv1(x)       
        x = F.relu(x)

        x = self.pool1(x)
        
        x = self.conv2(x)
        x = F.relu(x)

        x=self.pool2(x)


        # flatten => 32 x (32*26*26)
        x = x.flatten(start_dim = 1)

        # 32 x (32*26*26) => 32x128
        x = self.d1(x)
        x = F.relu(x)

        # logits => 32x10
        logits = self.d2(x)
        out = F.softmax(logits, dim=1)
        return out
```

```python
model = MyModel3()
x = torch.Tensor(4, 3, 64, 64) # batch, channel, rows, cols
out=model(x)
print(out.shape)
```
